```python
import pandas_ta as ta
import yfinance as yf
import matplotlib.pyplot as plt
import mplfinance as mpf
```

```python
ticker = yf.Ticker("0011.HK")
```

```python
ticker.info
```

```python
df = ticker.history(period="5y")
df
```

```python
df.ta.stoch( k=14, d=3, append=True)
df
```

```python
plots = [
    mpf.make_addplot(df['STOCHk_14_3_3'], panel=2),
    mpf.make_addplot(df['STOCHd_14_3_3'], panel=2)
]
mpf.plot(df, type="candle", volume=True, style="yahoo", addplot=plots, figscale=5)
```

```python
df["KD Diff"] = df["STOCHk_14_3_3"] - df["STOCHd_14_3_3"]
df["KD Diff rolling"] = df["KD Diff"].rolling(3).mean()
df.tail(10)
```

```python
volume_mean = df["Volume"].mean()
volume_mean
```

```python
def buy(price, principle):
    units = principle / price
    print(f"Buy {round(units, 2)} units with ${round(principle, 2)} at ${round(day_high, 2)}")
    return units

def sell(price, unit):
    value = price * unit
    print(f"Sell {round(units, 2)} units for ${round(value, 2)} at ${round(day_low, 2)}")
    return value
```

```python
initial_principle = 100_000
principle = initial_principle
units = 0
buy_price = 0

for index, day in df.iterrows():
    if abs(day["KD Diff rolling"]) < 10:
        if day["STOCHd_14_3_3"] < 20 and day["STOCHk_14_3_3"] < 20:
            if day["Volume"] > volume_mean:
                if principle > 0:
                    day_high = day["High"]
                    print(index)
                    units = buy(day_high, principle)
                    buy_price = day_high
                    principle = 0

        elif day["STOCHd_14_3_3"] > 80 and day["STOCHk_14_3_3"] > 80:
            if day["Volume"] > volume_mean:            
                if units > 0:
                    day_low = day["Low"]
                    if day_low > buy_price:
                        print(index)
                        principle = sell(day["Low"], units)

                        units = 0       

print()
print("Ending with", units)

# Sell all at last
if units > 0:
    principle = sell(day["Low"], units)

print()
# Final result %
print("$", round(principle, 2))

percentage = (principle-initial_principle) / initial_principle * 100
print(f"{round(percentage, 2)}%")
```
